```python
import pandas as pd
x = pd.read_csv("q1a.csv")
y = pd.read_csv("q1b.csv")
z = pd.read_csv("q1c.csv")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.title("Active Lowpass Filter")
plt.ylabel("Gain in dB")
plt.xlabel("Frequency $log_{10}(f)$")
plt.plot(np.log10(np.array(x['Freq.'])), 20*np.log10(np.array(x['Peak to Peak (mv)'])/1000))
plt.show()
```

```python
plt.title("Active Highpass Filter")
plt.ylabel("Gain in dB")
plt.xlabel("Frequency $log_{10}(f)$")
plt.plot(np.log10(np.array(y['Freq.'])), 20*np.log10(np.array(y['Peak to Peak (mv)'])/1000))
plt.show()
```

```python
plt.title("Multiple Feedback Active Bandpass")
plt.ylabel("Gain in dB")
plt.xlabel("Frequency $log_{10}(f)$")
plt.plot(np.log10(np.array(z['Freq.'])), 20*np.log10(np.array(z['Peak to Peak (mv)'])/1000))
plt.show()
```
