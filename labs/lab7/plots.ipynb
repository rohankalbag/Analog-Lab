```python
import pandas as pd
x = pd.read_csv("q1a.csv")
y = pd.read_csv("q1b.csv")
z = pd.read_csv("q1c.csv")
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
plt.title("Active Lowpass Filter")
plt.ylabel("Gain in dB")
plt.xlabel("Frequency $log_{10}(f)$")
plt.plot(np.log10(np.array(x['Frequency(Hz)'])), 20*np.log10(np.array(x['O/P Peak to Peak(mV)'])/1000), 'r-s')
plt.show()
```

```python
plt.title("Active Highpass Filter")
plt.ylabel("Gain in dB")
plt.xlabel("Frequency $log_{10}(f)$")
plt.plot(np.log10(np.array(y['Frequency(Hz)'])), 20*np.log10(np.array(y['O/P Peak to Peak(mV)'])/1000), 'r-s')
plt.show()
```

```python
plt.title("Multiple Feedback Active Bandpass")
plt.ylabel("Gain in dB")
plt.xlabel("Frequency $log_{10}(f)$")
plt.plot(np.log10(np.array(z['Frequency(Hz)'])), 20*np.log10(np.array(z['O/P Peak to Peak(mV)'])/1000), 'r-s')
plt.show()
```
